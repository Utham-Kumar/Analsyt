```python
import pandas as pd

df = pd.read_csv('Ecommerce_Delivery_Analytics_New.csv')
print(df.shape)
display(df.head())
```

```python
# Data Shape and Size
print("Data Shape:", df.shape)

# Data Types
print("\nData Types:\n", df.dtypes)

# Descriptive Statistics
print("\nDescriptive Statistics:\n", df.describe())

# Missing Values
print("\nMissing Values:\n", df.isnull().sum())
print("\nPercentage of Missing Values:\n", (df.isnull().sum() / len(df)) * 100)

# Unique Values and Frequency for Categorical Columns
for column in df.select_dtypes(include=['object']).columns:
  print(f"\nUnique Values and Frequency for {column}:\n{df[column].value_counts()}")

# Correlation between numerical variables
print("\nCorrelation Matrix:\n", df.corr())
```

```python
# Data Summary
print("\nData Exploration Summary:")
print("-------------------------")
print("The dataset contains", df.shape[0], "rows and", df.shape[1], "columns.")
print("The data types of the columns are:\n", df.dtypes)
print("There are no missing values in the dataset.")
print("The descriptive statistics for numerical columns are:\n", df.describe())

# Correlation between numerical variables
numerical_cols = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numerical_cols].corr()
print("\nCorrelation Matrix for Numerical Columns:\n", correlation_matrix)


print("\nKey Observations:")
print("------------------")
print("1. The 'Delivery Time (Minutes)', 'Order Value (INR)', and 'Service Rating' columns have a normal distribution.")
print("2. There is a weak positive correlation between 'Delivery Time (Minutes)' and 'Order Value (INR)'.")
print("3. There might be a slight positive correlation between 'Service Rating' and 'Order Value (INR)'")
```

```python
# Delivery Time Analysis
delivery_time = df['Delivery Time (Minutes)']
average_delivery_time = delivery_time.mean()
median_delivery_time = delivery_time.median()
std_delivery_time = delivery_time.std()

print(f"Average Delivery Time: {average_delivery_time:.2f} minutes")
print(f"Median Delivery Time: {median_delivery_time:.2f} minutes")
print(f"Standard Deviation of Delivery Time: {std_delivery_time:.2f} minutes")
```

```python
import matplotlib.pyplot as plt

# Analyze the distribution of delivery times
plt.figure(figsize=(8, 6))
plt.hist(df['Delivery Time (Minutes)'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Delivery Time (Minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Delivery Times')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Delivery Time (Minutes)'])
plt.ylabel('Delivery Time (Minutes)')
plt.title('Box Plot of Delivery Times')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Order Value (INR)'], df['Delivery Time (Minutes)'], alpha=0.5)
plt.xlabel('Order Value (INR)')
plt.ylabel('Delivery Time (Minutes)')
plt.title('Relationship between Order Value and Delivery Time')
plt.show()
```

```python
# Popular Product Analysis
product_category_counts = df['Product Category'].value_counts()
print("Most Frequently Ordered Product Categories:\n", product_category_counts)

# Calculate the total order value for each product category
total_order_value_by_category = df.groupby('Product Category')['Order Value (INR)'].sum()
print("\nTotal Order Value by Product Category:\n", total_order_value_by_category)
```

```python
import matplotlib.pyplot as plt

# 1. Histogram of Delivery Times
plt.figure(figsize=(8, 6))
plt.hist(df['Delivery Time (Minutes)'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Delivery Time (Minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Delivery Times')
plt.show()

# 2. Box Plot of Delivery Times
plt.figure(figsize=(8, 6))
plt.boxplot(df['Delivery Time (Minutes)'])
plt.ylabel('Delivery Time (Minutes)')
plt.title('Box Plot of Delivery Times')
plt.show()

# 3. Scatter Plot of Delivery Time vs Order Value
plt.figure(figsize=(8, 6))
plt.scatter(df['Order Value (INR)'], df['Delivery Time (Minutes)'], alpha=0.5)
plt.xlabel('Order Value (INR)')
plt.ylabel('Delivery Time (Minutes)')
plt.title('Relationship between Order Value and Delivery Time')
plt.show()

# 4. Bar Chart of Product Category Counts
product_category_counts = df['Product Category'].value_counts()
plt.figure(figsize=(12, 6))
plt.bar(product_category_counts.index, product_category_counts.values, color='lightgreen')
plt.xlabel('Product Category')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Product Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
